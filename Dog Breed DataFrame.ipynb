```python
import json
import pandas as pd
import requests
```

#### Make a GET request and get the JSON data

```python
url = 'https://dog.ceo/api/breeds/list/all'
response = requests.get(url)
json_data = response.json()
```

#### Write data into json file

```python
file_path = 'data.json'
with open(file_path, 'w') as json_file:
    json.dump(json_data, json_file)
```

#### Read data from the json File

```python
with open(file_path, 'r') as json_file:
    data = json.load(json_file)
```

#### Create a dataFrame

```python
#The pd.DataFrame.from_dict() function in pandas is used to create a DataFrame from a dictionary-like object.
#orient='index': This parameter specifies the orientation of the data in the dictionary. When orient is set to 'index', 
#it treats the dictionary keys as the row labels (index) in the resulting DataFrame.
```

```python
message = data['message']
df = pd.DataFrame.from_dict(message, orient='index')
```
